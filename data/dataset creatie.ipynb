```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv', index_col = 0)
```

```python
data.head(2)
```

```python
data['Length'] = data['content'].apply(lambda x: len(x.split()))
```

```python
data = data[data['Length'] < 1001]
```

```python
len(data)
```

```python
sample = data.sample(n=12)
```

```python
data = pd.concat([data, sample])
```

```python
data = data.drop(column, axis = 1)
```
